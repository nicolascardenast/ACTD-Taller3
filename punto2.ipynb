### Importación de los datos.

```python
# Importaciones.
import pandas as pd
import seaborn as sb

# Creación y visualización dataframes.
df = pd.read_csv('.\\real_estate_valuation_data_set.csv')
print(df.shape)
df.head()
```

### Estadísticas Descriptivas

```python
# Estadísticas descriptivas.
df[df.columns.difference(['No'])].describe()
```

### Histogramas y Diagramas de Dispersión

```python
# Excluir la columna "No".
sb.pairplot(df[df.columns.difference(['No'])])
```

### Matriz de Correlación

```python
# Matriz de correlación.
corr = df[df.columns.difference(['No'])].corr()
sb.heatmap(corr, cmap="Blues", annot=True)
```

### Exploración Bivariada

```python
# Primeras tres características.
sb.pairplot(df, x_vars=['X1 transaction date', 'X2 house age', 'X3 distance to the nearest MRT station'], y_vars='Y house price of unit area',
                         height=7, aspect=0.7, kind='reg')

# Últimas tres características.
sb.pairplot(df, x_vars=['X4 number of convenience stores',	'X5 latitude',	'X6 longitude'], y_vars='Y house price of unit area',
                         height=7, aspect=0.7, kind='reg')
```
