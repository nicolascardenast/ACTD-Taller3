```python
# Importaciones.
import pandas as pd
import seaborn as sns

# Creación y visualización dataframes.
df = pd.read_csv('.\\real_estate_valuation_data_set.csv')
print(df.shape)
df.head()
```

### **Explicación del significado de las variables**

1. **$X_1$, transaction date**: última vez que se realizó una transacción.
2. **$X_2$, house age**: edad de la casa en años.
3. **$X_3$, distance to the nearest MRT station**: distancia en metros a la estación de metro más cercana.
4. **$X_4$, number of convenience sotres**: número de tiendas.
5. **$X_5$, latitude**: latitud en grados.
6. **$X_6$, longitude**: longitud en grados.
7. **$Y$, house price of unit area**: precio de la casa por metro cuadrado en decenas de miles de dólares.

Además, se sabe que hay 414 inmuebles (observaciones) en total.
